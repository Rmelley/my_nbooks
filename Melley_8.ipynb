```python
import pandas as pd
import seaborn as sns
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.head()
```

```python
avocado.info()
```

```python
avocado['Date'] = pd.to_datetime(avocado['Date'])
avocado.info()
```

```python
bags = avocado.query('year == 2015' and 'type == "conventional"' and 'region == "Albany"')[['Date', 'Total Bags', 'Small Bags']]
bags.head()
```

```python
bags_melted = pd.melt(avocado, id_vars='Date',
                      value_vars=['Total Bags','Small Bags'],
                      var_name='Bags', value_name='Count')
bags_melted
```

```python
sns.relplot(data=bags_melted, kind='line', x='Date', y='Count', hue='Bags')
```

```python
avocado['Sales Volume'] = pd.qcut(avocado['Total Volume'], q=4,
labels=['Poor','Modest','Good','Excellent']) 

avocado
```

```python
sns.countplot(data=avocado, x='year', hue='Sales Volume')
```

```python
melted_avocado = pd.melt(avocado,id_vars=['year'],
                         value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'], 
                         var_name='Bags', 
                         value_name='Count') 
melted_avocado
```

```python
sns.relplot(data=melted_avocado, kind='line', x='year', y='Count', hue='Bags')
```
